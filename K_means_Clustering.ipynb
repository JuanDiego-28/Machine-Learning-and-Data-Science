```python
X = pd.get_dummies(df)
```

```python
X
```

```python
# scale the data 

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
scaled_X = scaler.fit_transform(X)
```

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters=2)
```

```python
cluster_labels = model.fit_predict(scaled_X)
```

```python
cluster_labels
```

```python
# assigning the labels 
X["Cluster"] = cluster_labels 
```

```python
X
```

```python
plt.figure(figsize=(12,4) ,dpi=200)
X.corr()["Cluster"].iloc[:-1].sort_values().plot(kind="bar")
```

```python
# lets measure best clustering

ssd = []

for k in range(2,10):
    model = KMeans(n_clusters=k)
    model.fit(scaled_X)


    ssd.append(model.inertia_)  ## SSD POINT --> CLUSTER CENTER
```

```python
ssd
```

```python
plt.figure(figsize=(12,4) ,dpi=200)
plt.plot(range(2,10) , ssd , "o--")
```

```python
pd.Series(ssd)  # calculate the diffferences with next and previous rows 
```

```python
pd.Series(ssd).diff()
```

#  Color quantization: Exerecise

```python
import matplotlib.image as mpimg
```

```python
image_array = mpimg.imread("../UNZIP_FOR_NOTEBOOKS_FINAL/DATA/palm_trees.jpg")
```

```python
image_array.shape # (H,W,C)
```

```python
plt.imshow(image_array)
```

```python
# flat the array

h,w,c = image_array.shape

image_2d = image_array.reshape(h*w, c)  # could use np.flatten
```

```python
model = KMeans(n_clusters=6)
```

```python
labels = model.fit_predict(image_2d)
```

```python
#labels
```

```python
rgb_codes = model.cluster_centers_.round(0).astype(int)  # round values
```

```python
rgb_codes
```

```python
q_image = np.reshape(rgb_codes[labels], (h,w,c))
```

```python
plt.imshow(q_image)
```
