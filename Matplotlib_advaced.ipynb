## The figure object

```python
import matplotlib.pyplot as plt

import numpy as np
```

```python
x = np.linspace(0,10,11)
v = x**4
```

```python
a = np.arange(0,10)
b= 2*a
```

```python
# store the object figure
fig = plt.figure()

# Large axes 
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(x,v)

axes1.set_ylim(0, 8000)
axes1.set_xlim(0 ,8)
axes1.set_title("the big one")

# small axes
axes2 = fig.add_axes([0.7,0.2,0.20,0.20])
axes2.plot(x,v)
axes2.set_ylim(1,500)
axes2.set_xlim(3,5)
axes2.set_title("the zoomed one")

plt.show()
```

```python
fig1 = plt.figure(figsize=(10,3) , dpi=100)

axes3 = fig1.add_axes([0,0,1,1])

axes3.plot(a,b)
```

## REMEBERING SUBPLOTS

```python
fig , axes = plt.subplots(nrows= 2 , ncols=2) 
axes[0][0].plot(x,v)
axes[0][1].plot(a,b)
axes[1][0].plot(x,v)
axes[1][1].plot(a,b)

fig.subplots_adjust(wspace=0.9)
```
