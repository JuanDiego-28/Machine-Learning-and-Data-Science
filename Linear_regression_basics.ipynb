```python
import numpy as np 
import pandas as pd 
import seaborn  as sns 
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv("/home/jdiego/Documentos/MACHINE_LEARNING_DATA_SCIENCE/UNZIP_FOR_NOTEBOOKS_FINAL/08-Linear-Regression-Models/Advertising.csv")
```

```python
df.head()
```

```python
df["total_spend"] = df["newspaper"] + df["TV"] + df["radio"]

df.head()
```

```python
sns.scatterplot(data=df , x="total_spend" , y="sales")
```

```python
# the regression line
sns.regplot(data=df , x="total_spend" , y="sales")
```

```python
X = df["total_spend"]
y= df["sales"]
```

```python
# using the polifits to y= mx + b 

a = np.polyfit(X,y,deg=1)
```

```python
potential_spend = np.linspace(0 , 500 , 100)

predicted_sales =  a[0]*potential_spend + a[1]
```

```python
sns.scatterplot(data=df , x="total_spend" , y="sales")
plt.plot(potential_spend , predicted_sales , color= "red")
```
