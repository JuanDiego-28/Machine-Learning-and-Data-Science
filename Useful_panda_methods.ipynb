## USEFUL METHODS

```python
# .apply() method ---> gives back values, not series 

def last_four(num):
    " this function returns last four digits of num"

    return int(str(num)[-4:])

data2["CC Number"].apply(last_four)    
```

##### lamda expression for multiple columns

```python
# lamda function

data2["total_bill"].apply(lambda num: num*2)
```

```python
def generous(total_bill , tip): 
    "define if a person gave a good tip or not"

    if tip/total_bill > 0.25: 
        return "generous"
    else:
        return "selfish"

data2[["total_bill" , "tip"]].apply(lambda data2: generous(data2["total_bill"] , data2["tip"]) , axis = 1 )
```

```python
#vectorized method from numpy
data2["Quality"] = np.vectorize(generous)(data2["total_bill"] , data2["tip"])

data2
```

```python
# to replace 
mymap = { "Female" :"F"  , "Male":"M"}   # also can use the .replace("Female" , "F") method 

data2["sex"].map(mymap)
```
