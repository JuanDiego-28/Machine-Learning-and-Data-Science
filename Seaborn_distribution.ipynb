```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
df = pd.read_csv("/home/jdiego/Documentos/MACHINE_LEARNING_DATA_SCIENCE/UNZIP_FOR_NOTEBOOKS_FINAL/05-Seaborn/dm_office_sales.csv")
df.head()
```

```python
# just the x axsis matters 
sns.rugplot(x="salary" , data=df , height=0.5)
```

```python
# this is a distribution plot 
#sns.set(style="dark")
sns.displot(data=df , x="salary" , bins=20 , kde=True)
```

```python
sns.histplot(data=df , x="salary" , bins=20 , kde=True)
```

```python
sns.kdeplot(data=df , x="salary")
```

# Categorical plots

```python
df["division"].value_counts()
```

```python
plt.figure(figsize=(10,4) , dpi=200)
sns.countplot(data=df , x="level of education" , hue="division" , palette="Set2")
```

```python
plt.figure(figsize=(10,4) , dpi=200)
sns.barplot(data=df , x="level of education" , y="salary" , estimator=np.mean , ci="sd")
```
