```python
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns 
import pandas as pd
```

```python
df = pd.read_csv("/home/jdiego/Documentos/MACHINE_LEARNING_DATA_SCIENCE/UNZIP_FOR_NOTEBOOKS_FINAL/05-Seaborn/StudentsPerformance.csv")

df.head()
```

```python
sns.boxplot(data=df , y="math score" , x="test preparation course")
```

```python
plt.figure(figsize=(10,4) , dpi=200)
sns.boxplot(data=df , y="reading score" , x="parental level of education" , hue="test preparation course")
plt.legend(bbox_to_anchor=(1.05,1))
```

```python
plt.figure(figsize=(10,7) , dpi=200)
sns.violinplot(data=df , x="reading score" , y="parental level of education" , hue="test preparation course" , split=True)
plt.legend(bbox_to_anchor=(1.05,1))
```

```python
plt.figure(figsize=(10,7) , dpi=200)
sns.violinplot(data=df , x="reading score" , y="parental level of education"  , split=True , inner="quartile")
plt.legend(bbox_to_anchor=(1.05,1))
```

```python
plt.figure(figsize=(10,7) , dpi=200)
sns.boxenplot( x="reading score" , y="parental level of education" , data=df )
plt.legend(bbox_to_anchor=(1.05,1))
```

```python
sns.pairplot(data=df , hue="gender" , corner=True)
```

# GRID plots

```python
sns.catplot(data=df , x="gender" , y="math score" , kind="box" , 
            row="lunch")
```

```python
sns.catplot(data=df , x="gender" , y="math score" , kind="box" , 
            row="lunch" , col="test preparation course")
```

```python
g = sns.PairGrid(df)

g = g.map_upper(sns.scatterplot)

g = g.map_diag(sns.histplot)

g= g.map_lower(sns.kdeplot)

```
