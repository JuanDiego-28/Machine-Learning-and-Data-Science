```python
import numpy as np 
import matplotlib.pyplot as plt 
import pandas as pd 
import seaborn as sns
```

```python
df = pd.read_csv("/home/jdiego/Documentos/MACHINE_LEARNING_DATA_SCIENCE/UNZIP_FOR_NOTEBOOKS_FINAL/08-Linear-Regression-Models/Advertising.csv")
```

```python
# split into x and y 

X = df.drop("sales" , axis=1)

y =df["sales"]
```

```python
from sklearn.preprocessing import PolynomialFeatures

poly_convertor = PolynomialFeatures(degree=3, include_bias=False)
```

```python
poly_features = poly_convertor.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(poly_features, y, test_size=0.3, random_state=101)
```

```python
# scaling data 

from sklearn.preprocessing import StandardScaler 

scaler = StandardScaler()
```

```python
scaler.fit(X_train)
```

```python
# scaled arrays 

X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

# L2 REGULARIZATION

```python
from sklearn.linear_model import Ridge
```

```python
ridge_model = Ridge(alpha=10)

ridge_model.fit(X_train , y_train )
```

```python
test_predictions = ridge_model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error , mean_squared_error 

rmae = mean_squared_error (y_test ,test_predictions)

rmae
```

```python
mae = mean_absolute_error (y_test, test_predictions)

mae
```

```python
# how to decide the best alpha 

from sklearn.linear_model import RidgeCV  # perform cross validation 
```

```python
ridge_cv = RidgeCV( alphas=(0.1 ,1.0 ,10.0) , scoring='neg_mean_absolute_error')
```

```python
ridge_cv.fit(X_train , y_train)
```

```python
ridge_cv.alpha_
```

```python
# scoring metrics

from sklearn.metrics._scorer import _SCORERS as SCORERS
#SCORERS.keys()
```

```python
test_predictions = ridge_cv.predict(X_test)  
```

```python
rmae = mean_squared_error (y_test ,test_predictions)

np.sqrt(rmae)
```

```python
mae = mean_absolute_error (y_test, test_predictions)

mae
```

# L1 REGULARIZATION

```python
from sklearn.linear_model import LassoCV
```

```python
lasso_cv = LassoCV(eps=0.001, n_alphas=100 , cv=5 , max_iter=1000000)
```

```python
lasso_cv.fit(X_train, y_train)
```

```python
lasso_cv.alpha_
```

```python
test_predictions =lasso_cv.predict(X_test)
```

```python
rmae = mean_squared_error (y_test ,test_predictions)

np.sqrt(rmae)
```

```python
mae = mean_absolute_error (y_test, test_predictions)

mae
```
